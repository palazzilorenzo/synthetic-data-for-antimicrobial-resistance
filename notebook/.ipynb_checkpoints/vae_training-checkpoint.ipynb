```python
# Import data (Escherichia coli spectra) into dataframe with Pandas

path_to_data = '../VAE/real_data/Escherichia_coli_VAE'
extension = '.txt'

df = pd.read_csv(path_to_data + extension, sep='\t', header=0)
df.head()
```

### Training step

```python
# import training data
path_to_data = '../VAE/real_data/train/Escherichia_coli_VAE_train'
extension = '.csv'

df_train = pd.read_csv(path_to_data + extension, sep='\t', header=0)
df_train.head()
```

```python
# Remove the 'Unnamed: 0' column

df_train = df_train.drop(['Unnamed: 0'],  axis = 1)
df_train.shape
```

```python
# preparing data to train the VAE

train = df_train.to_numpy()

train = np.expand_dims(train, -1).astype("float32")
print(train.shape)
```

To create the Variational Autoencoder, call the Vae() method and pass it as argument the latent space dimension. 
For example, below, we create six VAEs with six different latent space dimension. In vae_results notebook we can see that the quality of synthetic data increases with increasing of latent space dimension.

```python
# create VAEs

vae_2 = Vae(2)
vae_4 = Vae(4)
vae_8 = Vae(8)
vae_16 = Vae(16)
vae_32 = Vae(32)
vae_64 = Vae(64)
```

```python
# This callback will stop the training when there is no improvement in
# the loss for ten consecutive epochs.

callback = EarlyStopping(monitor='loss', patience=10)
```

```python
history_2 = vae_2.fit(train, epochs=150, batch_size=9, callbacks=[callback])
print("Number of epochs run", len(history_2.history['loss']))

# save weights
vae_2.save_weights('../VAE/weights/vae_2_weights.h5')
# save history on file
with open('../VAE/history/history_vae_2.json', 'w') as f:
    json.dump(history_2.history, f)
```

```python
history_4 = vae_4.fit(train, epochs=150, batch_size=9, callbacks=[callback])
print("Number of epochs run", len(history_4.history['loss']))

# save weights
vae_4.save_weights('../VAE/weights/vae_4_weights.h5')

# save history on file
with open('../VAE/history/history_vae_4.json', 'w') as f:
    json.dump(history_4.history, f)
```
